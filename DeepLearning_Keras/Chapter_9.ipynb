## 9장. 결론

### 핵심 개념 리뷰

* AI를 위한 여러 방법 
    * 인공지능 : 인지 과정을 자동화하기 위한 모든 방법(사고의 자동화) 
        * 엑셀 스프레드 시트 ~ 휴머노이드 로봇까지 저수준~고수준 모두 해당 
    * 머신러닝 : 훈련 데이터를 사용하여 자동으로 모델을 생성하는 AI의 특정 하위 분야
        * 학습 : 데이터를 모델(프로그램)으로 바꾸는 과정
    * 딥러닝 : 머신 러닝의 여러 종류 중 하나
        * 기하학적 변환 함수들이 번갈아 가며 연속적으로 길게 연결된 모델
        * 연산들이 층이라는 모듈을 구성.층을 쌓아올려서 딥러닝 모델을 구성. 층의 그래프
        * 층은 훈련하는 동안 학습되는 가중치 파라미터를 가짐 -> 좋은 가중치 값을 찾는 것이 목표

* 머신 러닝 분야에서 딥러닝이 특별한 이유 
    * 기술적인 성공
        * 많은 기업에 딥러닝이 비즈니스 가치를 제공하고 있음
            * 사람 수준의 음성 인식, 이미지 분류, 스마트 비서, 크게 향상된 기계 번역
        * 더 이상 기술이 발전되지 않더라도 현재 알고리즘을 가능한 모든 문제에 적용만해도 큰 혁신을 이룰 것

* 딥러닝에 대하여
    * 단순함 : 경사 하강법으로 충분히 큰 모수 모델을 충분히 많은 샘플에서 훈련하는 것만으로도 높은 성과를 얻을 수 있음
        * 벡터 : 모델 '입력'과 '타깃'이 먼저 '벡터'로 바뀌어야함
        * 모델 : 각 층은 데이터에 간단한 기하학적 변환을 수행한 뒤 통과 시킴
            * 모델은 '층'을 연결하여 복잡한 '기하학적 변환'을 구성
        * 변환이 입력 공간을 타깃 공간으로 한번에 하나의 포인트 씩 매핑
            * 모델이 얼마나 잘 작동하는지를 기반으로 '가중치 파라미터'를 반복적으로 업데이트
            * '기하학적 변환'의 핵심 특징은 '미분 가능'해야 함
                * 경사 하강법으로 파라미터를 학습하기 위해서는 '미분 가능'해야 함
                    * 입력에서 출력으로 바뀌는 '기하학적 변환'이 부드럽고 연속적이어야 함
    * 딥러닝 모델 : 고차원 데이터의 복잡한 매니폴드를 펼치는 수학 장치
        * 데이터가 가진 의미를 '벡터'와 '기하학적 공간'으로 변환한 후 한 공간에서 다른 공간으로 매핑하는 복잡한 '기하학적 변환'을 점진적으로 학습
    
    * 모든 것이 핵심 아이디어에서 출발
        * '의미'는 어떤 '두 가지 사이'(언어에 있는 단어 사이, 이미지에 있는 픽셀 사이 등)'관계'에서 유도
            * 관계 : 거리 함수로 측정
        * 벡터 공간 : 컴퓨터가 계산하기에 매우 효율적인 환경
        
    * 딥러닝(신경망)의 호칭
        * 층 기반 표현 학습(layer representations learning)
        * 계층적 표현 학습(hierarchical representations learning)
        * 심층 미분 모델(deep differentiable model) : 연속하여 기하학적 공간을 조작
        * 기하학적 변환 연결(chained geometric transform)
        
* 핵심 기술 :딥러닝의 발전 요인
    * 알고리즘 혁신
    * 지각에 관련된 많은 양의 데이터(인터넷의 성장, 무어의 법칙)
    * 고성능 병렬 컴퓨터 하드웨어를 값싸게 사용 가능(Nvida GPU)

* 일반적인 머신 러닝 작업 흐름
    1. 문제를 정의 : 사용할 데이터, 예측 대상, 데이터 수집 방법 등
    2. 목표 달성을 측정하기 위해 신뢰할 수 있는 방법을 찾음 : 예측 정확도, 여러 문제 영역에 특화된 정교한 지표가 필요
    3. 모델을 평가하기 위해 사용할 검증 과정을 준비 : test_set, validation_set, test_set
        * 시계열 예측의 경우 검증 데이터와 테스트 데이터는 시간 순서상 훈련 데이터 뒤에 와야함 
    4. 데이터를 벡터화하고 신경망에 잘 맞는 형태로 전처리(정규화 등)
    5. 상식 수준의 기본 모델보다 나은 첫 번째 모델을 생성 한 뒤 확인
    6. 하이퍼파라미터 튜닝, 규제를 추가해서 모델을 개선
        * 테스트 데이터나 훈련 데이터를 사용하지 않고 검증 데이터의 성능만 사용하여 조정
    7. 하이퍼파라미터 튜닝을 하면 검증 세트에 오버피팅 될 수 있음 : test_set을 따로 떼어 놓음

* 주요 네트워크 구조 : 특정 입력 형식을 의미 
    * 완전 연결 네트워크
    * 합성곱 네트워크
    * 순환 네트워크
