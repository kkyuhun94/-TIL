### 2.Matrix Algebra

> 2.1 Matrix Operations

* Theorem 1 : Let $A,B,C$ be same size, $r,s$ be scalars
    * $a. A+B=B+A$
    * $b. (A+B)+C=A+(B+C)$
    * $c. A+0=A$
    * $d. r(A+B)=rA+rB$
    * $e. (r+s)A=rA+sA$
    * $f. r(sA)=(rs)A$

* Matrix multiplication
    * $A(Bx)=(AB)x$
    * In general, $AB \neq BA$ 
    * $A,B$는 서로 commute : $AB=BA$를 성립한다.

* Theorem 2 : Let $A$ be $mxn$, $B,C$ be $nxk$, $r$ be scalar, $I$ be identitiy matrix
    * $a. A(BC)=AB(C)$
    * $b. A(B+C)=AB+AC$
    * $c. (B+C)A=BA+CA$
    * $d. r(AB)=(rA)B=A(rB)$
    * $e. {I}_mA=A=A{I}_n$
    
    
* Transpose of a matrix
    * $A$가 $mxn$ matrix일 때,  transpose of $A(A^T)$는 $nxm$
    * Transpose : 대각선 대칭으로 row와 column의 자리 바꿈
* Theorem 3 : Let $A,B$ sum,product에 적합
    * $a. (A^T)^T=A$
    * $b. (A+B)^T=A^T+B^T$
    * $c.$ For any scalar $r, (rA)^T=rA^T$
    * $d. (AB)^T=B^TA^T$
> 2.2 The Inverse of a Matrix

* Invertible matrix $A$: Matrix $A$가 Inverse Matrix(역행렬) $A^{-1}$을 가짐
    * inverse matrix is unique.
    * invertible = nonsingular
    * not invertible = singular

* Theorem 4. Let $A=\begin{pmatrix} a & b\\ c & d \end{pmatrix}$
    * if $ad-bc \neq 0$
    * then $A$ is invertible,
    $$ A^{-1}=\frac{1}{ad-bc}\begin{pmatrix} d & {-b}\\ -c & a \end{pmatrix}$$
    * determinant : $detA=ad-bc$

* Theorem 5. Let $A$ is an invertible $nxn$ matrix, then for each $b$ in $R^n$
    * the equation $Ax=b$ has the unique solution $x$=$A^{-1}b$
    
* Theorem 6. If A is an invertible matrix, then $A^(-1)$ is invertible and $(A^{-1})^{-1}=A$
    * If $A,B$ are $n x n$ invertible matrices, $(AB)^{-1} = B^{-1}A{-1}$
    * $(A^T)^{-1}=(A^{-1})^T$

* Elementary matrix : single elementary row operation on an identity matrix.
    * identity matrix에서 row operation(replace,scailing)을 1회
    * 만약 elementary row operation이 $mxn$matrix $A$에 실행된다면,
        * $EA$라고 표현. 이때 E는 $mxm$, 단위행렬$I_m$에 같은 row operation
        * 각 elementary matrix E는 invertible. 
            * $E^{-1}$은 I에서 E로 변한 과정을 거꾸로 I에 적용하면 구할 수 있다

* Theorem 7
    * $nxn$ matrix $A$ is invertible = row equivalent to $I_n$
    * 어떤 elementary row operation의 연속($E_1E_2E_3..E_p$)이 $A$를 $I_n$으로 transform한다. 
        * $EA = I$
        * $EI = A^-1$ = $E$
    
* Algorithm for finding ${A}^{-1}$
    * Augmented matrix $[A$  $I]$ ~ $[I$ $C]$
    * $C = A^{-1}$
    
> 2.3 Character of Invertible Matrices

* Theorem8. The invertible matrix theorem : Let A $nxn$ matrix
    * equivalent!
    * $a. A$ is invertible
    * $b. CA=I$인 $nxn$ matrix $C$가 존재한다.
    * $c. Ax=0$은 오직 trivial solution을 갖는다
    * $d. A$은 pivot position을 $n$개 갖고 있다.
    * $e. A$는 indentity $nxn$ matrix $I$와 row equivalent
    * $f. AD=I$를 만족하는 $nxn$ matrix $D$가 존재
    * $g. Ax=b$는 최소한 1개의 solution을 $b$ in $R^n$을 갖는다.
    * $h.$columns of $A$ span $R^n$
    * $i.$ linear transformation $A$가 $R^n$on to$R^n$ 
    * $j. A$의 컬럼은 linearly independent set : $Ax=0$이 trivial solution을 갖는다. 
    * $k.$ linear transformation $A$가 one to one
    * $l. A^T$ is invertible 
    
* Theorem 9 : Invertible linear transformation
    * $T:R^n \to R^n $가 linear transformation, $A$는 standard matrix일 때,
    * $T$가 invertible 이면 $A$도 invertible
    * $S$라는 transformation이 존재하고 $A^{-1}$가 standard matrix 

> 2.4 Partitioned Matrices 

* Partitioned matrix
    * $A = \begin{pmatrix} 1 & 0 & {-3} & 5 & {-4} & 1&2\\ {-8} & 9 & 8 & 6 & 7 & -1&3 \\0&{-5}&1&5&7&2&0 \end{pmatrix}$
    * $A = \begin{pmatrix} {A_{11}} & {A_{12}}&{A_{13}}\\ {A_{21}}& {A_{22}}&{A_{23}} \end{pmatrix}$
    * $A+B , rA$

* Multiplication of partitioned matrices
    * partition matrix A,B를 곱하려면 A의 컬럼과 B의 row를 같은 크기로 partition해야 한다.
    
* Theorem 10. Column-row Expansion of $AB$
    * if $A$ is $m$ x $n$, $B$ is $n$ x $p$
    * $AB=col_1(A)row_1(B)+...+col_n(A)row_n(B) = [Ab_1 $.....$Ab_p]$
    

* Inverse of partitioned matrix
    * $p$x$p$ matrix $A = \begin{pmatrix} A_{11} & A_{12} \\ 0 & A_{22} \end{pmatrix}$
    * A는 block upper triangular, Diagonal은 반드시 invertible,즉 A는 Invertible
    * $B$ = $A^{-1}$ = $ \begin{pmatrix} {A^{-1}}_{11} & -{A^{-1}}_{11}A_{12}{A^{-1}}_{11} \\ 0 & {A^{-1}}_{22} \end{pmatrix}$
