```python
# class 
# man.py
# 쉘에서 python man.py와 같은 결과 

class Man :
    def __init__(self, name) :   # 생성자
        self.name = name
        print("Initialized!")
    
    def hello(self) :        # 메서드 1
        print("Hello " + self.name + "!")
    
    def goodbye(self) :      # 메서드 2
        print("Good-bye " + self.name + "!")
        
m = Man("David")
m.hello()
m.goodbye()
```
